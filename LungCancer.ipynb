```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
lung_data["GENDER"] = lung_data["GENDER"].map({"M": 1, "F": 2})
lung_data["LUNG_CANCER"] = lung_data["LUNG_CANCER"].map({"YES": 1, "NO": 0})
#encoding to 1 & 0 for applying our model
```

```python
# lung_data["GENDER"]
lung_data["LUNG_CANCER"]
```

```python
x = lung_data.iloc[:, 0:-1]
y = lung_data.iloc[:, -1]
```

```python
from sklearn.linear_model import LogisticRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
# x_train.head()
y_test.head()
```

## Logistic Regression

```python
#Fitting simple linear regression to the training test
Model1 = LogisticRegression()
Model1.fit(x_train, y_train)
#Predicting the test set results
prediction1 = Model1.predict(x_test)
```

```python
prediction1
#1 - Yes (Cancer) , 0 - No Cancer
```

```python
confusion_matrix(y_test,prediction1)
```

```python
accuracy_score(y_test,prediction1)
```

```python
probs = Model1.predict_proba(x_test)
precision_score(y_test, prediction1, average = None)

# Precision for No Cancer (class 0) = 0.75
#  Out of all the people the model predicted No Cancer, 75% actually had no cancer.

# Precision for Has Cancer (class 1) = 0.8947
#  Out of all the people the model predicted Has Cancer, about 89.5% truly had cancer.
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, prediction1)
precision = precision_score(y_test, prediction1)
recall = recall_score(y_test, prediction1)
f1 = f1_score(y_test, prediction1)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1)

#very important , depending on the need , we may prefer recall or precision or accuracy , f1 score is balance between recall and precision
#accuracy -> To find out the percentage of correct predictions out of all predictions.

#precision -> Precision = Correct Positive Predictions / Total Predicted Positives
# eg: if precision is 0.5 then Out of all people "your model" said have cancer, only 50% actually do.

#recall -> Correct Positive Predictions / Total Actual Positives
# eg: if recall is 0.75(15/20) then Out of the 20 people who "actually" have cancer, your model caught 15  so 75% recall.

#f1 score ->To get a balanced measure when you care about both precision and recall (especially in unbalanced data).
#f1 score Example: If you want both fewer false positives and to catch as many true cases as possible  F1 score helps balance that.
```

```python
cm = confusion_matrix(y_true = y_test, y_pred = prediction1)
#plot_confusion_matrix(cm,level,title = "confusion_matrix")
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```
